> 1. 

\begin{gather*}
    \mu = \frac{6 + 3 + 3 + 2 + 2}{5} \\
    \mu = \frac{16}{5} \\
    \mu = 3.2 \\ \\

    \sigma^2 = \frac{(6 - 3.2)^2 + (3 - 3.2)^2 + (3 - 3.2)^2 + (2 - 3.2)^2 + (2 - 3.2)^2}{5} \\
    \sigma^2 = \frac{(2.8)^2 + (-0.2)^2 + (-0.2)^2 + (-1.2)^2 + (-1.2)^2}{5} \\

    \sigma^2 = 2.16 \\

\end{gather*}

> 3. 

\begin{gather*}
    \mu = \frac{1 + 2 + 1}{3} \\
    \mu = \frac{4}{3} \\
    \mu = 1.33 \\ \\

    \sigma^2 = \frac{(2.33)^2 + (3.33)^2 + (2.33)^2}{3} \\
    \sigma^2 = \frac{5.43 + 11.09 + 5.43}{3} \\
    \sigma^2 = \frac{21.95}{3} \\
    \sigma^2 = 7.32 \\
    \sigma = \sqrt{7.32} \\
    \sigma = 2.71
\end{gather*}

The formula in getting a single variance is $x - \mu$ and we shift the dataset by $4$ the new formula will be $(x + 4) - (\mu + 4)$ since shifting also affects the mean. Now, we can simplify the new formula to $x - \mu$ which is the same as the original formula so no change has occurred at all.
