## Quiz 20: Covariance

> 1. 

\begin{gather*}
    \sigma_x = \frac{2 + 3 + 4 + 5 + 6}{5} = \frac{20}{5} = 4 \\
    \sigma_y = \frac{3 + 5 + 7 + 9 + 11}{5} = \frac{35}{5} = 7 \\ \\

    Cov[x,y] = = \frac{[-2 \cdot (-4)] + [-1 \cdot (-2)] + (0 \cdot 0) + (1 \cdot 2) + (2 \cdot 4)}{4} \\
    Cov[x,y] = \frac{8 + 2 + 2 + 8}{4} \\
    Cov[x,y] = \frac{20}{4} \\
    Cov[x,y] = 5
\end{gather*}
