## Exercise 6: Covariance

> 1. A bakery records sales and number of customers for a sample of hours
throughout the week. Calculate the covariance of customers and sales.



\begin{gather*}
    \sigma_{\text{Customers}} = \frac{4 + 7 + 12 + 2 + 3 + 9 + 15}{7} = \frac{52}{7} = 7.43 \\
    \sigma_{\text{Sales}} = \frac{45.75 + 36 + 58.5 + 20 + 15.8 + 39.95 + 123.45}{7} = \frac{339.45}{7} = 48.49 \\ \\

    Cov[\text{Customers}, \text{Sales}] = \frac{(-3.43)(-2.74) + (-0.43)(-12.49) + (4.57)(10.01) + (-5.43)(-28.49) + (-4.43)(-32.69) + (1.57)(-8.54) + (7.57)(74.96)}{6} \\
    Cov[\text{Customers}, \text{Sales}] = \frac{9.4 + 5.37 + 45.75 + 154.7 + 144.82 - 13.41 + 567.45}{6} \\
    Cov[\text{Customers}, \text{Sales}] = \frac{914.08}{6} \\
    Cov[\text{Customers}, \text{Sales}] = 152.35
\end{gather*}
