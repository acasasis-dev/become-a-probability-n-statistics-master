## Exercise 2: The Addition Rule and Union vs. Intersection

> 1. Given the probabilities $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.05$, what is
$P(A \cup B)$? Are $A$ and $B$ mutually exclusive events? Why or why not?

\begin{gather*} 
        P(\text{A}) = \frac{0.3}{1} \\
        P(\text{B}) = \frac{0.6}{1} \\
        P(\text{A} \cap \text{B}) = \frac{0.05}{1} \\
\\
        P(\text{A} \cup \text{B}) = \frac{0.3}{1} + \frac{0.6}{1} - \frac{0.05}{1} \\
        P(\text{A} \cup \text{B}) = \frac{0.85}{1} \\
        P(\text{A} \cup \text{B}) = 0.85 \\
        P(\text{A} \cup \text{B}) = 85.0\%
\end{gather*}

$A$ and $B$ are **NOT** mutually exclusive because $P(A \cap B)$ has a value which is not $0$.

trips up the hill. What is the probability that Jack and Jill both fell down on
any particular trip, and what is the probability that only Jack fell down or
only Jill fell down on any particular trip?



\begin{gather*}

        P(\text{Jack} \cup \text{Jill}) = 22.97 \\

\end{gather*}

> 5. Every student at a certain high school needs to choose exactly one
fine arts elective. The frequency table shows the enrollment of electives
for all students. Are the events junior and architecture mutually
exclusive? What is the probability that a student is taking architecture and
a junior? What is the probability that a student is a junior or is taking
architecture?



The events "junior" and "architecture" are NOT mutually exclusive as they have $45$ overlapping probabilities.

        P(\text{Junior} \cup \text{Architecture}) = 296.91 \\

\end{gather*}

> 6. James tosses a coin and rolls a six-sided die. What is the sample
space for this situation? What is the probability the coin lands on heads
and the die lands on a $2$ or a $3$?

The sample space would be $2 \times 6 = 12$.

\begin{gather*}
    P(\text{heads} \ \cap 2 \ \text{or} \ 3) = \frac{1}{12} + \frac{1}{2} = \frac{2}{12} = \frac{1}{6}
\end{gather*}
