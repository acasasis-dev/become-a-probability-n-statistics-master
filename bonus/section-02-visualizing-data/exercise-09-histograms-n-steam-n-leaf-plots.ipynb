## Exercise 9: Histograms and Steam and Leaf Plots

> 1. A doctor recorded the weight of all the babies that visited her clinic
last week. How many babies weighed no more than $24$ pounds?



> 2. The stem plot shows the number of clothing pieces on each rack at a
clothing store. Create a histogram from the stem plot, and use buckets of
size $10$.



> 3. Is it possible to create a stem-and-leaf plot from a histogram? Why or
why not?

> 4. A company mails out packets of dog treat samples based on a
consumer’s previous dog food purchases. How many times did the
company mail a packet of $11 − 15$ treats?



> 5. A company mails out packets of dog treat samples based on a
consumer’s previous dog food purchases. How many packets of dog treat
samples did the company give out?



> 6. Create a stem-and-leaf chart from the list of student test scores.
