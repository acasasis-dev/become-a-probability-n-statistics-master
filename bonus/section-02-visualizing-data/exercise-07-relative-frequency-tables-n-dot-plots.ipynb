## Exercise 7: Relative Frequency Table

```python
import matplotlib.pyplot as plt
import numpy as np
```

> 1. Blake is surveying students in his class (made up of juniors and
seniors) about whether or not they play video games on a daily basis.
What type of relative frequency table is shown? Finish filling in the table.



Total-relative frequency.

```python
columns = ['Play at least one video game daily', 'Don\'t play any video games daily', 'Total']
rows = ['Junior', 'Senior', 'Total']
data = [
	['23%', '52%', '75%'],
	['11%', '14%', '25%'],
	['34%', '66%', '100%']
]

fig, ax = plt.subplots()
ax.axis('off')

table = plt.table(
	colLabels=columns,
	rowLabels=rows,
	cellText=data,
	loc='center',
	cellLoc='center',
)


plt.show()
```
