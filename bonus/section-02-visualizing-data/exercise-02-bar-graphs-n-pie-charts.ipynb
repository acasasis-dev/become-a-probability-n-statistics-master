## Exercise 02: Bar Graphs and Pie Charts

> 1. Both the bar graph and the table have missing information about the
number of bones each dog consumed at doggie daycare. Use the graph
and table together to fill in the missing pieces.





```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.DataFrame({
	'Dog': ['T-Bone', 'Pepper', 'Cookie', 'Patches', 'Matty', 'Davy'],
	'Number of Bones': [3, 1, 1, 5, 4, 2]
})

# Create a figure and axis
fig, ax = plt.subplots()

# Hide axes
ax.axis('off')

plt.title("Number of Bones Each Dog Consumed At Doggie Daycare")
# Create the table
table = plt.table(
    cellText=data.values,
    colLabels=data.columns,
    cellLoc='center',
    loc='center'
)

table.scale(1, 2)
plt.show()

plt.figure(figsize=(5, 4))
sns.barplot(data=data, x=data.columns[0], y=data.columns[1])

plt.grid(axis='y', zorder=2)
plt.tight_layout()
plt.show()
```

> 2. Eric’s class went on a trip to an orchard. At the end of the trip they
counted how many pieces of fruit came from each type of tree and
graphed it in the bar graph shown below. Use the bar graph to create a pie
chart of the data.



```python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Plums', 'Pears', 'Peaches']
sizes = [22, 33, 17, 28]

# Plot
plt.figure(figsize=(4,4))  # Circle shape
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pieces of Fruit From Each Type of Tree')
plt.axis('equal')  # Equal aspect ratio = circle
plt.tight_layout()
plt.show()
```

> 4. The pie chart shows how many ice cream cones of each flavor were
sold. Assuming $280$ total ice cream cones were sold in August, convert the
pie chart to a bar graph.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
	'Flavor': ['Strawberry', 'Chocolate', 'Vanilla', 'Peanut Butter'],
	'Cones Sold': [60, 80, 100, 40]
})

plt.figure(figsize=(5, 4))
sns.barplot(data=data, x=data.columns[0], y=data.columns[1])

plt.grid(axis='y', zorder=2)
plt.tight_layout()
plt.show()
```
