> 1. Create a comparison bar graph for the two-way table.



```python
categories = ['Fish', 'Cat', 'Dog', 'Other']
first_grade = [8, 15, 7, 9]
second_grade = [13, 10, 12, 5]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, first_grade, width, label='First Grade')
rects2 = ax.bar(x + width/2, second_grade, width, label='Second Grade')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total')
ax.set_title('Comparison of favorite pet of 1st and 2nd graders')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

> 2. A pizza parlor wants to know if the age range of their customers
affects pizza preferences. The pizza parlor asks each customer two
questions:

1. Which type of pizza is your favorite: pepperoni, cheese, supreme
or veggie?
2. What is your age range: Under $18$, or $18$ and over?

The results of the survey are as follows:
- Of the $50$ customers who prefer pepperoni pizza, $25$ are under $18$.
- Of the $20$ customers who prefer cheese pizza, $18$ are under $18$.
- Of the $30$ customers who prefer supreme pizza, $24$ are over $18$.
- Of the $25$ customers who prefer veggie pizza, $19$ are over $18$.

Which type of table, one-way or two-way, can be created from the data
that the pizza parlor is collecting? Create the best type of frequency table
for the data.

The type of table that is suited for this is two-way as the data answers 2 questions.

```python
columns = ['Pepperoni', 'Cheese', 'Supreme', 'Veggie', 'Total']
rows = ['Under 18', '18 and over', 'Total']
data = [
    [25, 18, 6, 6, 55],
	[25, 2, 24, 19, 70],
	[50, 20, 30, 25, 125]
]

# Create a figure and axis
fig, ax = plt.subplots()

# Hide the main axes
ax.axis('off')

# Add a table
table = plt.table(
    cellText=data,
    rowLabels=rows,
    colLabels=columns,  # skip 'Name' since it's used as row labels
    loc='center',
    cellLoc='center'
)

# Adjust layout
table.scale(1, 1.5)  # scale width, height
plt.show()
```

> 4. Decide whether a comparison bar graph or comparison line graph
would be better at displaying the data in the two-way table, then create
the graph.



```python
grade = ['Pre-School', 'First', 'Second', 'Third']
walk = [1, 5, 10, 8]
school_bus = [10, 12, 22, 33]
day_care_vehicle = [20, 14, 5, 3]
carpool = [26, 19, 15, 10]

x = np.arange(len(grade))
width = 0.2

fig, ax = plt.subplots()
ax.bar(x - width * 1.5, walk, width, label='Walk')
ax.bar(x - width/2, school_bus, width, label='School Bus')
ax.bar(x + width/2, day_care_vehicle, width, label='Day Care Vehicle')
ax.bar(x + width * 1.5, carpool, width, label='Carpool')
# Add labels, title, and legend
ax.set_xlabel('Grade in School')
ax.set_ylabel('Method of transportation')
ax.set_title('Method of transportation per Grade in School')
ax.set_xticks(x)
ax.set_xticklabels(grade)
ax.legend()

# Display the plot
plt.show()
```

> 5. Eric creates a survey asking students who ate a snack in the morning
between classes if they felt sleepy or not. Given his survey results below,
create a two-way data table for Eric’s survey.

`

```python
columns = ['Feel Sleepy? Yes', 'Feel Sleepy? No']
rows = ['Ate A Snack? Yes', 'Ate A Snack? No']
data = [
	['5', '2'],
	['3', '5']
]

fig, ax = plt.subplots()
ax.axis('off')

table = plt.table(
	rowLabels=rows,
	colLabels=columns,
	cellText=data,
	loc='center',
	cellLoc='center',
)

plt.show()
```
