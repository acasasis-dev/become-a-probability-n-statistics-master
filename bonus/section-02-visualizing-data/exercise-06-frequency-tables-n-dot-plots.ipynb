## Exercise 6: Frequency Tables and Dot Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

> 1. The frequency table shows the number of seed packets sold by each
child during a pre-school fundraiser. Create a dot plot from the frequency
table.



```python
# Sample data
data = (
	[1] * 5 + # Ivan
    [2] * 6 + # Stacy
    [3] * 3 + # Vanessa
    [4] * 8 + # Josh
    [5] * 5 + # Jamie
    [6] * 7 + # Kelly
    [7] * 10 + # Billy
    [8] * 5 + # Cassie
    [9] * 7 + # Time
    [10] * 3 # Kate
)

# Calculate histogram to get counts for each bin
bins = np.arange(12) - 0.5 # Bins for integer values
hist, edges = np.histogram(data, bins=bins)

# Generate plotting coordinates
x_coords = []
y_coords = []
for i, count in enumerate(hist):
    value = int(edges[i] + 0.5) # Get the integer value for the bin
    for j in range(1, count + 1):
        x_coords.append(value)
        y_coords.append(j)

# Create the dot plot
plt.figure(figsize=(10, 6))
plt.scatter(x_coords, y_coords, s=50, alpha=0.7) # s for marker size, alpha for transparency
print()
plt.xticks(np.arange(0, 10)) # Set x-axis ticks for clarity
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Dot Plot of Sample Data")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

> 2. The dot plot shows the age of people who bought a bag of kale at a
grocery store. Create a frequency table from the dot plot.



```python
columns = ['Number of Bags Bought']
rows = ['18 to 34', '35 to 64', '65+']
data = [
	[8], [10], [3]
]

fig, ax = plt.subplots()
ax.axis('off')
table = plt.table(
	colLabels=columns,
	rowLabels=rows,
	cellText=data,
	cellLoc='center',
	loc='center'
)

plt.show()
```

> 5. Kevin watches birds from his window and records what kind he sees.
Create a dot plot from the data.
