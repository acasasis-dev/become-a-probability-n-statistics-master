> 1. If the range of the data set is $36$ and we want to divide it into $5$ class
intervals, which of the following would be the most appropriate class
width?

Divide $36$ to $5$ and round the results up

\begin{gather*}
    \frac{36}{5} = 7.2 = 8
\end{gather*}

The most appropriate class width is $8$
